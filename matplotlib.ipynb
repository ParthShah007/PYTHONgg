```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basics
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x, y)
plt.title("Y=XÂ² GRAPH")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
```

```python
fig = plt.figure(figsize=(10, 5), dpi=100)
axis1 = fig.add_axes([0.1, 0.1, 1, 1])
axis1.plot(x, y)
axis2 = fig.add_axes([0.3, 0.4 , 0.5, 0.5])
axis2.plot(x,y)
plt.savefig("myplot.png")
```
