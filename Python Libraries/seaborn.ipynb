```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Distribution plots for quantitative data, 
Categorical plot for quantitative data

```python
df = sns.load_dataset("tips")
df
```

```python
df['size'].unique()
 # bins = number of bins
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], bins = 20, kde = True) # kde = kernel density estimate
plt.subplot(1,2,2)
sns.histplot(df["tip"], bins = 20, kde = True)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "kde") # More hex ,reg, kde, scatter
```

```python
sns.pairplot(df, hue='size', palette='rainbow') # hue = categorical variable to color by
```

```python
sns.rugplot(df) # Rug plot shows distribution of data along an axis
```

Categorical data ke graphs


```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x = df['sex'], hue = df['smoker']) # Count plot shows the count of observations in each categorical bin using bars
```

```python
sns.barplot(x = df['sex'], y=df['total_bill'], hue= df['smoker']) # Bar plot shows the mean of the total bill
```

```python
import numpy as np
sns.barplot(x = df['sex'], y=df['total_bill'], estimator = np.sum)
```

```python
sns.boxplot(x = "tip", y = "day", data = df, palette='rainbow') 
```

```python
sns.violinplot(x = "day", y = "total_bill", data = df, palette='rainbow') # Violin plot shows the distribution of the data across different categories
```

```python
sns .stripplot(x = "day", y = "total_bill", data = df, palette='rainbow') # Strip plot shows the distribution of the data across different categories with jitter
```

```python
sns.swarmplot(x = "day", y = "total_bill", data = df, palette='rainbow') # Swarm plot shows the distribution of the data across different categories without overlap
```

```python
sns.violinplot(x = "tip", y = "day", data = df, palette='rainbow') # Violin plot shows the distribution of the data across different categories
sns.swarmplot(x = "tip", y = "day", data = df, palette='rainbow') # Swarm plot shows the distribution of the data across different categories without overlap
```

MATRIXPLOTS

```python
#Heat map
flights = sns.load_dataset("flights")
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot = True)
```

```python
# Cluster Map
sns.clustermap(tipscorr.corr(), annot = True) # Cluster map shows the correlation between the variables in a clustered heatmap
```

```python
pvt_flight = flights.pivot_table(index ='month', columns='year', values='passengers')
pvt_flight
```

```python
sns.heatmap(pvt_flight)
```
