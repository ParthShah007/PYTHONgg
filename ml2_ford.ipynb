```python
x = df.drop(columns=['price'], axis=1)
y = df['price']
x
```

```python
df.columns
```

```python
# One-hot encoding categorical variables
x_one_encode = pd.get_dummies(x, columns = ['transmission', 'fuelType', 'model'], drop_first=True)
x_one_encode = x_one_encode.astype(int)
x_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
columns = ['transmission', 'fuelType', 'model']
xlabel = x
for i in columns:
    xlabel[i] = encode.fit_transform(x[i])
xlabel
```

```python
from sklearn.preprocessing import StandardScaler
cols = ['year', 'mileage', 'engineSize', 'tax', 'mpg']
scaler = StandardScaler()
x_one_encode[cols] = scaler.fit_transform(x_one_encode[cols])
x_one_encode
```

```python
xlabel.columns
```

```python
xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
xlabel
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(xlabel, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = x_test.shape[0]
m = x_test.shape[1]
r2_adj = 1 - (1 - r2) * (n - 1) / (n - m - 1)
r2_adj
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
r2
```
